# Computing the $\pi$ number

```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
```

> The Circle equation is : $x^2 + y^2 = r^2$ where *r* is the radius.

So the function of the semicircle with radius *$r$* is : $f(x)=\sqrt{r^2-x^2}$.

```python
def f(x) : 
    x = min(x , 1) #to handle endpoint error
    return mt.sqrt(1 - x**2)
```

```python
x=np.arange(-1 , 1 + 0.01 , 0.01)
plt.plot(x , np.array(list(map(f , x))))
```

The surface of a circle with radius 1 ($r$ = 1) is $\pi$ so the semicircle's surface would be $\frac{\pi}{2}$.

Our approach is to divide the semicircle to some rectangles with width of $n$ and sum of them would be it's surface.

> It's actually : $\int_{-1}^{1} f(x) = \frac{\pi}{2} $

```python
def surface(x1,x2,n):
    return (f(x1) + f(x2))/2 * 2/n # just a good estimation    
```

```python
def divide(n) : 
    s = 0
    for i in np.arange(-1 , 1 - 2/n , 2/n) :
        s += surface(i , i + 2/n , n)
    return s 
```

So now $divide(n)$ will divide the [-1 , 1] range to $n$ pieces and calculate the surface of that subjected rectangle and add them up!

> As you know when $n$ rises, our estimate is going to get better and the best result occur when $n \rightarrow \infty$

> !Note that the result is going to be $\frac{\pi}{2}$ so we will times it to 2.

```python
divide(1000) * 2
```

```python
mt.pi
```

That wasn't a bad estimate but let's check the error plot . 

```python
x = np.arange(100 , 10000 , 50) 
plt.plot(x , abs(np.array(list(map(divide , x)))*2 - mt.pi))
```

And now the actual function : 

```python
plt.plot(x , np.array(list(map(divide , x)))*2)
```

Let's see the best estimate that we can achieve in 1 minute of computation : 

```python
divide(10000000) * 2
```
